## Dataset Description

The Iris dataset contains 150 records of iris flowers belonging to three species:
Iris-setosa, Iris-versicolor, and Iris-virginica.

Each record includes:
- Sepal length (cm)
- Sepal width (cm)
- Petal length (cm)
- Petal width (cm)

The dataset is commonly used for data analysis and classification tasks.

## Data Cleaning and Preparation

The dataset was checked for missing values and duplicate rows.
No missing or duplicate values were found, so no further cleaning was required.

```python
# Cell 9: Data cleaning checks & NumPy operations
# Check for missing values
print("Missing values in each column:\n", df.isnull().sum())

# Check for duplicates
print("Number of duplicate rows:", df.duplicated().sum())

# Convert a column to NumPy array
sepal_length = df['SepalLengthCm'].values

# Basic statistics using NumPy
print("\nSepal Length Stats using NumPy:")
print("Mean:", np.mean(sepal_length))
print("Median:", np.median(sepal_length))
print("Standard Deviation:", np.std(sepal_length))
```

```python
# Create a new figure
plt.figure()

# Scatter plot of sepal length vs sepal width
sns.scatterplot(
    x='SepalLengthCm',
    y='SepalWidthCm',
    hue='Species',
    data=df
)

# Add plot title
plt.title('Sepal Length vs Sepal Width')

# Show the plot
plt.show()
```

```python
plt.figure()
plt.hist(df['PetalLengthCm'], bins=20)
plt.xlabel('Petal Length')
plt.ylabel('Frequency')
plt.title('Petal Length Distribution')
plt.show()
```

```python
plt.figure()
sns.boxplot(
    x='Species',
    y='SepalLengthCm',
    data=df
)
plt.title('Sepal Length by Species')
plt.show()
```

## Exploratory Data Analysis (EDA)

- Scatter plot shows clear separation between species based on sepal dimensions.
- Histogram shows petal length distribution is right-skewed.
- Box plot shows variation of sepal length among different species.
